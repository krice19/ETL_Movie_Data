## Iteration 1


```python
# appling a list comp to filter out any entities that don't have a director & imbd link
#the list comp filters through the raw JSON
wiki_movies = [movie for movie in wiki_movies_raw 
               if ('Director' in movie or 'Directed by' in movie) and 'imdb_link' in movie]
```

```python
# checking how many rows of data 
len(wiki_movies)
# we started at 7311, so this list comp filtered out around 231 rows 
```

```python
# putting wiki_movies into a dataframe to inspect
wiki_movies_df = pd.DataFrame(wiki_movies)
wiki_movies_df.head()
```

## Iteration 2

```python
# adding to the list comp to remove any entites that contain a value in the "number of episodes" column, to remove any tv shows from te data
wiki_movies = [movie for movie in wiki_movies_raw 
               if ('Director' in movie or 'Directed by' in movie) and 'imdb_link' in movie
              and 'No. of episodes' not in movie]
```

```python
len(wiki_movies)
```

```python
# putting wiki_movies into a dataframe to inspect
wiki_movies_df = pd.DataFrame(wiki_movies)
wiki_movies_df.head()
```
